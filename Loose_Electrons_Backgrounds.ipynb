```python
for sel in ['REG', 'RSE', 'TLE'] :
    for fs in ['Zpe', 'Zpm'] :
        legend_position = (0.2, 0.7, 0.6, 0.85)
        if sel != 'REG' and fs == 'Zpm' : continue
        if sel == 'TLE' : legend_position = (0.6, 0.7, 0.9, 0.85)
        print 'selection ', sel, ' for state ', fs
        plot_FR(sel, '', fs)
```
